Flipkart E-commerce Data Analysis using Python



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("flipkart_com-ecommerce_sample.csv")
df.head(6)
```

```python
df.isnull().sum()
```

```python
print("\n Dataset Info:")
df.info()
print("\n Describe Dataset:")
df.describe()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['brand'].fillna("unknown",inplace=True)
```

```python
df['discount_percentage'] = (df['retail_price'] - df['discounted_price']) / df['retail_price'] * 100
```

```python
top_brands = df['brand'].value_counts().head(10)
```

```python
top_category = df['product_category_tree'].value_counts().head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title('Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_category.index, y=top_category.values)
x=top_category.values
y=top_category.index
plt.title('Top 10 Category')
plt.xlabel('Count')
plt.ylabel('Category')
plt.xticks(rotation=50)
plt.show()
```

```python
df['brand'] = df['brand'].fillna('unknown')
top5_brands = df['brand'].value_counts().head(5)
plt.figure(figsize=(10,6))
plt.pie(top5_brands, labels=top5_brands.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Brands Share')
plt.show()
```

```python
scatter_data = df[['retail_price','discounted_price']].dropna()
plt.figure(figsize=(10,8))
sns.scatterplot(x='retail_price', y='discounted_price', data=scatter_data)
plt.title('Retail Price VS Discounted Price')
plt.xlabel('Retail Price')
plt.ylabel('Discounted Price')
plt.show()
```
